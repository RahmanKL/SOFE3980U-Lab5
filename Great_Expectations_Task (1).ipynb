##5. Set Up Great Expectations Context and Data Source

```python
context = gx.get_context()
data_source = context.data_sources.add_pandas("pandas")
data_asset = data_source.add_dataframe_asset(name="labels_data")
```

##6. Define and Create a Data Batch

```python
batch_definition = data_asset.add_batch_definition_whole_dataframe("batch_definition")
batch = batch_definition.get_batch(batch_parameters={"dataframe": df})
```

### Expectation 1

```python
# Expectation 1: Each image filename in Occluded_Image_view should be unique
# A duplicate filename would mean the same image was used for two different timestamps,
# which indicates a data recording error.
expectation_1 = gx.expectations.ExpectColumnValuesToBeUnique(
    column="Occluded_Image_view"
)
```

### Validate Data Against Expectation 1

```python
result_1 = batch.validate(expectation_1)
print(result_1)
```

### Expectation 2

```python
# Expectation 2: Pedestrian bounding box X coordinates (top-left) must be within
# valid image pixel range [0, 1920]. Values of 854 appear as sentinel "no detection"
# placeholders, so we use a wide but bounded range to catch truly out-of-range entries.
expectation_2 = gx.expectations.ExpectColumnValuesToBeBetween(
    column="pedestrianLocationX_TopLeft",
    min_value=0,
    max_value=1920
)
```

### Validate Data Against Expectation 2

```python
result_2 = batch.validate(expectation_2)
print(result_2)
```

### Expectation 3

```python
# Expectation 3: Ground_Truth_View filenames must match the naming convention C_NNN.png
# where NNN is a zero-padded 3-digit number. Any deviation indicates a corrupted or
# misnamed label file that could break downstream processing pipelines.
expectation_3 = gx.expectations.ExpectColumnValuesToMatchRegex(
    column="Ground_Truth_View",
    regex=r"^C_\d{3}\.png$"
)
```

### Validate Data Against Expectation 3

```python
result_3 = batch.validate(expectation_3)
print(result_3)
```
